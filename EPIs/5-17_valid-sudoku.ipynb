# Valid Sudoku

Determine if a 9x9 Sudoku board is valid.  Only the filled cells need to be validated according to the following rules:

- Each row must contain the digits `1-9` without repetition
- Each column must contain the digits `1-9` without repetitionn
- Each of the 9 `3x3` sub-boxes of the grid must contain the digits `1-9`

Note

1. A Sudoku board (partial filled) could be valid but is not necessarily solveable.
2. Only the filled cells need to be validated according to the rules

## Example 1

```
Input: board = 
[
["5","3",".",".","7",".",".",".","."],
["6",".",".","1","9","5",".",".","."],
[".","9","8",".",".",".",".","6","."],
["8",".",".",".","6",".",".",".","3"],

[".",".",".",".","8",".",".","7","9"]
]
Output: true
```

````python
def hasDuplicates(items):
    return False if len(items) != len(set(items)) else True

def hasCorrectNumbers(items):
    for i, item in enumerate(items):
        if item < 1 or item > 9:
            return False
    return True

def removeBlank(items):
    return [i for i in items if i != "."]

def isValid(items):
    return not hasDuplicates(items) and hasCorrectNumbers(items)

def isValidSudoku(board):
    num_of_rows = len(board)
    
    for i in range(num_of_rows):
        row = removeBlank(board[i][:])
        column = removeBlank([j[i] for j in board])
        box = removeBlank(board)

        if not isValid(row) or not isValid(column) or not isValid(box):
            return False
        
    return True
        
        


print(isValidSudoku([
    ["8","3",".",".","7",".",".",".","."],
    ["6",".",".","1","9","5",".",".","."],
    [".","9","8",".",".",".",".","6","."],
    ["8",".",".",".","6",".",".",".","3"],
    ["4",".",".","8",".","3",".",".","1"],
    ["7",".",".",".","2",".",".",".","6"],
    [".","6",".",".",".",".","2","8","."],
    [".",".",".","4","1","9",".",".","5"],
    [".",".",".",".","8",".",".","7","9"]]))
````
